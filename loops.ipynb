```python
def KSA(key):
    key_length = len(key)
    S = list(range(256))
    j = 0

    for i in range(256):
        j = (j + S[i] + key[i % key_length]) % 256
        S[i], S[j] = S[j], S[i]  # Swap

    return S


key = [ord(c) for c in "SecretKey"]  # Convert string to ASCII values
S = KSA(key)
print(S)
```
