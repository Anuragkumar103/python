two side of a rectangle 

```python
def check_sides_equal(side1, side2):
  """
  Checks if two given sides of a rectangle are equal.

  Args:
    side1: The length of the first side.
    side2: The length of the second side.

  Returns:
    True if the sides are equal, False otherwise.
  """
  return side1 == side2

length = float(input("Enter the length of the rectangle: "))
width = float(input("Enter the width of the rectangle: "))

if check_sides_equal(length, width):
  print("The two sides are equal. This is a square.")
else:
  print("The two sides are not equal. This is a rectangle (but not a square).")

side_a = 5
side_b = 5
side_c = 10
side_d = 10

if check_sides_equal(side_a, side_b):
  print("Side A and Side B are equal.")
else:
  print("Side A and Side B are not equal.")

if check_sides_equal(side_a, side_c):
  print("Side A and Side C are equal.")
else:
  print("Side A and Side C are not equal.")
```
